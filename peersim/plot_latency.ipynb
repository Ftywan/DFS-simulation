```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename1 = "exp/1000nodes0.01drop0minD25maxD0.3singleReq0.3add0.3delete-latency.csv"
df1 = pd.read_csv(filename1)

filename2 = "exp/2000nodes0.01drop0minD25maxD0.3singleReq0.3add0.3delete-latency.csv"
df2 = pd.read_csv(filename2)

filename3 = "exp/1000nodes0.01drop0minD50maxD0.3singleReq0.3add0.3delete20K-request.csv"
```

```python
variable = "network"
```

```python
plt.hist(df1[df1['type']=='ADD']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="1000_ADD")
plt.hist(df1[df1['type']=='DELETE']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="1000_DELETE")
plt.hist(df1[df1['type']=='RETRIEVE']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="1000_RETRIEVE")
plt.hist(df2[df2['type']=='ADD']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="2000_ADD")
plt.hist(df2[df2['type']=='DELETE']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="2000_DELETE")
plt.hist(df2[df2['type']=='RETRIEVE']['duration'].values.tolist(), bins=75, cumulative=True, histtype='step', label="2000_RETRIEVE")

plt.xlabel("Latency")
plt.ylabel("Count")
plt.legend(loc='upper left')
plt.savefig(f'{variable}cdf.jpg', dpi=500)
plt.show()
```

```python
df1 = df1.groupby('type')['duration'].agg(sum="sum", mean="mean", median="median")
```

```python
df2 = df2.groupby('type')['duration'].agg(sum="sum", mean="mean", median="median")
```

```python
types = ['ADD', 'DELETE', 'RETRIEVE']

plt.bar([x -0.2 for x in np.arange(len(types))], df1['median'].values.tolist()[0], width=0.4, label='1000 nodes')
plt.bar([x + 0.2 for x in np.arange(len(types))], df2['median'].values.tolist()[0], width=0.4, label='2000 nodes')
plt.xticks(np.arange(len(types)), types)
plt.xlabel('Request Type')
plt.ylabel('Median Delay')
plt.legend()
plt.savefig(f'{variable}bar.jpg', dpi=500)
plt.show()
```
